## Exercise 1: Square the Numbers

You’re given a list of numbers.  
Write a `for` loop that prints the **square** of each number.

````python
numbers = [2, 4, 6, 8]

# Your code here
for number in numbers:
  print(number ** 2)
````

## Exercise 2: Calculate Total Cost

You’re given a list of item prices.  
Write a `for` loop that adds up the total cost and prints the **final total**.


````python
item_prices = [5.99, 3.49, 12.99, 2.99]

# Your code here
total = 0
for item_price in item_prices:
  final_total += item_price

print("Total cost", final_total)
````

<details>
<summary>Solution - click to reveal</summary>

```
item_prices = [5.99, 3.49, 12.99, 2.99]

total = 0
for price in item_prices:
    total += price

print("Total cost:", total)

```

## Exercise 3: Countdown Timer

Use a `for` loop with `range()` to count down from **5 to 1**,  
then print `"Blast off!"`.

````python
# Your code here
for timer in range(5, 0, -1):
  print(timer)
print("Blast off!")
````

<details>
<summary>Solution - click to reveal</summary>

```
for number in range(5, 0, -1):
    print(number)

print("Blast off!")

```

</details>


## Exercise 4: Print Birthday Messages

You’re given a list of names.  
Write a `for` loop that prints a birthday message for each person.

````python
friends = ["Ava", "Ben", "Cleo"]

# Your code here
for friend in friends:
  print("Happy Birthday", friend+ "!")
````

<details>
<summary>Solution - click to reveal</summary>

```
friends = ["Ava", "Ben", "Cleo"]

```

## Exercise 5: Multiples of 3

Use `range()` with a custom step to print all multiples of 3  
from 3 up to (and including) 30.


````python
# Your code here
for multiple in range(3, 31):
  if multiple % 3 == 0:
    print(multiple)

````
